# 2  Data Cleaning

```python
df_clean=pd.read_excel("/content/Customer Call List.xlsx")
df_clean
```

```python
df_clean=df_clean.drop_duplicates()
df_clean
```

```python
df_clean=df_clean.drop(columns="Not_Useful_Column")
df_clean
```

```python
# Clean up any unwanted characters ("/..._")
df_clean["Last_Name"]=df_clean["Last_Name"].str.strip("/..._")
df_clean
```

```python
# Convert 'Phone_Number' column to string type to enable string slicing
df_clean["Phone_Number"] = df_clean["Phone_Number"].astype(str)


#  Remove non-numeric characters
df_clean["Phone_Number"] = df_clean["Phone_Number"].str.replace(r'[^0-9]', '', regex=True)


# Apply the formatting only if the string has enough characters
df_clean["Phone_Number"] = df_clean["Phone_Number"].apply(lambda x: x[0:3] + '-' + x[3:6] + '-' + x[6:10] if len(x) >= 10 else x)
df_clean
```

```python
# Splits the Address column into three new columns
df_clean[["Street_Address", "State", "Zip_Code"]] = df_clean['Address'].str.split(',', n=2, expand=True)
df_clean
```

```python
df_clean["Paying Customer"] = df_clean["Paying Customer"].str.replace('Yes','Y')
df_clean["Paying Customer"] = df_clean["Paying Customer"].str.replace('No','N')
df_clean
```

```python
df_clean["Do_Not_Contact"] = df_clean["Do_Not_Contact"].str.replace('Yes','Y')
df_clean["Do_Not_Contact"] = df_clean["Do_Not_Contact"].str.replace('No','N')
df_clean
```

```python
df_clean=df_clean.replace('N/a','')
df_clean=df_clean.fillna('')
df_clean
```
